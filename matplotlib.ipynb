```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


data = {
    "gamma": [0.99,0.99,0.75,0.75,0.50,0.50,0.25,0.10,0.10],
    "theta": ["1e-6","1e-3","1e-6","1e-3","1e-1","1e-3","1e-3","1e-6","1e-3"],
    "reward": [-68.33,-39.73,-84.27,-84.27,-38.60,-89.17,-79.13,-79.13,-41.77],
    "congestion": [72,26,83,83,15,88,73,70,19]
}

df = pd.DataFrame(data)
```

```python
plt.figure()

plt.plot(df["gamma"], df["reward"], marker='o')

plt.xlabel("Gamma (Discount Factor)")
plt.ylabel("Total Reward")
plt.title("Effect of Gamma on Total Reward")

plt.grid()

plt.show()

plt.figure()
```

```python
plt.plot(df["gamma"], df["congestion"], marker='o')

plt.xlabel("Gamma (Discount Factor)")
plt.ylabel("Congestion Events")
plt.title("Effect of Gamma on Traffic Congestion")

plt.grid()

plt.show()
```

```python
pivot = df.pivot(index="gamma", columns="theta", values="reward")

plt.figure()

sns.heatmap(pivot, annot=True)

plt.title("Gamma vs Theta vs Reward")

plt.show()
```

```python
avg_ns = [4.47,3.62,10.29,10.29,6.45,11.19,10.94,10.74,6.99]
avg_ew = [13.03,7.70,12.59,12.59,5.13,12.56,10.40,10.60,5.54]

plt.figure()

plt.plot(df["gamma"], avg_ns, marker='o', label="NS Cars")
plt.plot(df["gamma"], avg_ew, marker='o', label="EW Cars")

plt.xlabel("Gamma")
plt.ylabel("Average Cars Waiting")
plt.title("Average Waiting Cars vs Gamma")

plt.legend()
plt.grid()

plt.show()
```
