```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

data = {
    "gamma": [0.75, 0.50, 0.50, 0.25, 0.10, 0.10],
    "theta": ["1e-3", "1e-1", "1e-3", "1e-3", "1e-3", "1e-6"],
    "reward": [-45.07, -64.23, -38.53, -77.63, -32.13, -45.07],
    "congestion": [19, 45, 13, 60, 13, 19]
}

df = pd.DataFrame(data)

print(df)
```

```python
plt.figure()

plt.plot(df["gamma"], df["reward"], marker='o')

plt.xlabel("Gamma (Discount Factor)")
plt.ylabel("Total Reward")
plt.title("Effect of Gamma on Total Reward")

plt.grid()

plt.show()

plt.figure()
```

```python
plt.plot(df["gamma"], df["congestion"], marker='o')

plt.xlabel("Gamma (Discount Factor)")
plt.ylabel("Congestion Events")
plt.title("Effect of Gamma on Traffic Congestion")

plt.grid()

plt.show()
```

```python
pivot = df.pivot(index="gamma", columns="theta", values="reward")

plt.figure()

sns.heatmap(pivot, annot=True)

plt.title("Gamma vs Theta vs Reward")

plt.show()
```

```python

avg_ns = [6.59, 7.19, 5.67, 8.00, 4.66, 6.44]
avg_ew = [6.93, 9.98, 5.89, 11.09, 5.28, 7.08]

plt.figure()

plt.plot(df["gamma"], avg_ns, marker='o', label="NS Cars")
plt.plot(df["gamma"], avg_ew, marker='o', label="EW Cars")

plt.xlabel("Gamma")
plt.ylabel("Average Cars Waiting")
plt.title("Average Waiting Cars vs Gamma (Policy Iteration)")

plt.legend()
plt.grid()

plt.show()
```
